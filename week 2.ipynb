```python
# Notebook about analysing the customers data 
```

```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# 1- What is the number of customers?
```

```python
customers = pd.read_csv('customers1.csv')
customers
```

```python
Number_customers=customers.shape[0]
Number_customers
```

```python
# 2- perecentage of customers per geography
```

```python
geo_percentage= customers['Geography'].value_counts(normalize=True)*100
geo_percentage
```

```python
plt.figure(figsize=(4,4))
plt.pie(geo_percentage, labels=geo_percentage.index, autopct='%1.1f%%', colors=['lightcoral','lightgreen','lightblue'])
plt.title('percentage of customers per geography')
plt.show()
```

```python
# 3- Average of Credit Score per geography
```

```python
avg_credit_score= customers.groupby('Geography')['CreditScore'].mean()
avg_credit_score
```

```python
sns.barplot(x=avg_credit_score.index,y=avg_credit_score.values, color='lightblue', width=0.5, edgecolor='red')
plt.title(' Average Credit Score per geography')
plt.ylabel(' avg credit score')
plt.show()
```

```python
# 4- Average of Credit Score per age
```

```python
avg_credit_score_age= customers.groupby('Age')['CreditScore'].mean()
avg_credit_score_age
```

```python
plt.scatter(x=avg_credit_score_age.index,y=avg_credit_score_age.values, color='red', marker='*',s= 80, alpha=1)
plt.title(' Average Credit Score per age')
plt.ylabel(' avg credit score')
plt.show()
```

```python
# 5- Total of Estimated salary  per age
```

```python
Total_Estimatedsalary_age= customers.groupby('Age')['EstimatedSalary'].sum()
Total_Estimatedsalary_age
```

```python
plt.scatter(x=Total_Estimatedsalary_age.index,y=Total_Estimatedsalary_age.values, color='gold', marker='+',s= 80, alpha=1)
plt.title(' Total of Estimated Salary per age')
plt.ylabel(' Sum of Estimated Salary')
plt.show()
```

```python
plt.hist(Total_Estimatedsalary_age, color='gold')
plt.title(' Histogram of total estimated salary')
plt.show()
```

```python
# 6- Average of Estimated salary  per geography
```

```python
Average_Estimatedsalary_geography= customers.groupby('Geography')['EstimatedSalary'].mean()
Average_Estimatedsalary_geography
```

```python
sns.barplot(x=Average_Estimatedsalary_geography.index,y=Average_Estimatedsalary_geography.values, color='lightcoral', width=0.5, edgecolor='red')
plt.title(' Average Estimated Salary per geography')
plt.ylabel(' avg estimated salary')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.pie(Average_Estimatedsalary_geography, labels=Average_Estimatedsalary_geography.index, autopct='%1.1f%%', colors=['lightcoral','lightgreen','lightblue'])
plt.title('Average of Estimated salary per Geography')
plt.show()
```

```python
# 7- Average of Estimated salary  per gender
```

```python
Average_Estimatedsalary_gender= customers.groupby('Gender')['EstimatedSalary'].mean()
Average_Estimatedsalary_gender
```

```python
sns.barplot(x=Average_Estimatedsalary_gender.index,y=Average_Estimatedsalary_gender.values, color='lightgreen', width=0.5, edgecolor='green')
plt.title(' Average Estimated Salary per gender')
plt.ylabel(' avg estimated salary')
plt.show()
```

```python
# 8- Average of Credit Score  per gender
```

```python
Average_CreditScore_gender= customers.groupby('Gender')['CreditScore'].mean()
Average_CreditScore_gender
```

```python
plt.figure(figsize=(4,4))
plt.pie(Average_CreditScore_gender, labels=Average_CreditScore_gender.index, autopct='%1.1f%%', colors=['lightcoral','lightgreen','lightblue'])
plt.title('Average of Credit Score Per Gender')
plt.show()
```

```python
#9- Number Of Customers per Geography
```

```python
Number_customers_Geography= customers.groupby('Geography')['CustomerId'].count()
Number_customers_Geography
```

```python
sns.barplot(x=Number_customers_Geography.index,y=Number_customers_Geography.values, color='gold', width=0.3, edgecolor='red')
plt.title(' Number of customers per geography')
plt.ylabel(' number of customers')
plt.show()
```

```python
#10- Number Of Females and Males per Geography
```

```python
Number_gender_Geography= customers.groupby(['Geography','Gender']).size().unstack()
Number_gender_Geography
```

```python
plt.figure(figsize=(10,6))
Number_gender_Geography.plot(kind='bar', stacked=True, color=['skyblue','lightpink'])
plt.title('Number of Males and Females per geography')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.legend(title='Gender')
plt.show()
```
